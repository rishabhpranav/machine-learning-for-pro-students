```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
matches = pd.read_csv('ipl/matches.csv')
deliveries = pd.read_csv('ipl/deliveries.csv')
```

```python
matches['season'].value_counts()
```

```python
plt.hist(matches['season'],rwidth=0.9)
```

```python
sns.countplot(x='season',data=matches)
```

```python
plt.figure(figsize=(10,6))
# sns.countplot(x='winner',data=matches)
# plt.xticks(rotation='vertical')

matches['winner'].value_counts().plot.bar(width=0.9)
```

```python
sns.countplot(x='toss_decision',data=matches,hue='toss_decision')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='season',data=matches,hue='toss_decision')
```

```python
six_df = deliveries[deliveries['batsman_runs'] == 6]
```

```python
plt.figure(figsize=(10,6))
ax = six_df['batsman'].value_counts().head(10).plot.bar(width=0.9,color=sns.color_palette('inferno'))

for patches in ax.patches:
    ax.annotate(format(patches.get_height()), (patches.get_x() + 0.3, patches.get_height() + 4))
```

```python
max_runs = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
ax = max_runs.plot.bar(width=0.9,color=sns.color_palette('inferno'))
for patches in ax.patches:
    ax.annotate(format(patches.get_height()), (patches.get_x() + 0.3, patches.get_height() + 60))
```

```python
winner = matches['toss_decision'].value_counts()
```

```python
labels = winner.index
```

```python
wins = winner.values
```

```python
plt.figure(figsize=(10,6))
plt.pie(wins,labels=labels,startangle=90,autopct='%1.2f%%',explode=(0.2,0), shadow=True)
plt.legend()
plt.show()
```
